```python

import numpy as np
r1=np.array([2.5,8,-5])
r2=np.array([-1,6,3])
cross_pro = np.cross(r1, r2)
mag_cross_pro = np.linalg.norm(cross_pro)
norm_r1 = np.linalg.norm(r1)
norm_r2 = np.linalg.norm(r2)

sin_theta = mag_cross_pro / (norm_r1 * norm_r2)
theta_rad_cross = np.arcsin(sin_theta)
finaltheta = np.degrees(theta_rad_cross)

print("The angle between r1 and r2 in degrees is:", finaltheta)
```

```python
import numpy as np
def mat_multi_numpy(Alist,Blist):
   # Args:
   #    Alist(list[[int/float]]):the first matrix
   #    Blist(list[[int/float]]):the second matrix
   # Returns:
   #    numpy.ndarray: The resulting matrix
   A=np.array(Alist)
   B=np.array(Blist)
   C=A@B
   return C
```

```python
# Example usage:
matrix_a = [[1, 2], [3, 4]]
matrix_b = [[5, 6], [7, 8]]
result_matrix = mat_multi_numpy(matrix_a, matrix_b)
print("Result of matrix multiplication:")
print(result_matrix)
```

```python
import sympy

# Define symbolic variables
x, y, z = sympy.symbols('x y z')

# Define the vector field G
G = sympy.Matrix([y**2 + z, -x + z**2, x**2 + y])

# Calculate the curl
curl_G = sympy.Matrix([
    sympy.diff(G[2], y) - sympy.diff(G[1], z),  # i component
    sympy.diff(G[0], z) - sympy.diff(G[2], x),  # j component
    sympy.diff(G[1], x) - sympy.diff(G[0], y)   # k component
])

# Simplify the curl
curl_G = sympy.simplify(curl_G)

# Print the curl
print("Curl of G:")
sympy.pprint(curl_G)

# You can also get individual components:
curl_i = curl_G[0]
curl_j = curl_G[1]
curl_k = curl_G[2]
```

```python
import sympy

# Define symbolic variables
x, y, z = sympy.symbols('x y z')

# Define the scalar function phi
phi = x**2 + y + y * z**2

# Calculate the gradient
gradient_phi = sympy.Matrix([
    sympy.diff(phi, x),  # Partial derivative with respect to x
    sympy.diff(phi, y),  # Partial derivative with respect to y
    sympy.diff(phi, z)   # Partial derivative with respect to z
])

# Simplify the gradient
gradient_phi = sympy.simplify(gradient_phi)

# Print the gradient
print("Gradient of phi:")
sympy.pprint(gradient_phi)

# Evaluate the gradient at the point (1, 2, -1)
point = {x: 1, y: 2, z: -1}
gradient_at_point = gradient_phi.subs(point)
print("\nGradient at point (1, 2, -1):")
sympy.pprint(gradient_at_point)
```

```python
import sympy

# Define symbolic variables
x, y, z = sympy.symbols('x y z')

# Define the vector field G (example)
G = sympy.Matrix([y**2 + z, -x + z**2, x**2 + y])

# Calculate the divergence
divergence_G = sympy.diff(G[0], x) + sympy.diff(G[1], y) + sympy.diff(G[2], z)
divergence_G = sympy.simplify(divergence_G)

# Print the divergence
print("Divergence of G:")
sympy.pprint(divergence_G)

# Evaluate the divergence at the point (1, 2, -1)
point = {x: 1, y: 2, z: -1}
divergence_at_point = divergence_G.subs(point)
print("\nDivergence at point (1, 2, -1):")
sympy.pprint(divergence_at_point)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def projectile_motion_simulation():

    v0 = float(input("Enter initial velocity (v0) in m/s: "))
    theta_deg = float(input("Enter launch angle (theta) in degrees: "))
    y0 = float(input("Enter initial vertical position (y0) in meters: "))
    x0 = float(input("Enter initial horizontal position (x0) in meters: "))
    theta_rad = np.radians(theta_deg)


    g = 9.81


    vox = v0 * np.cos(theta_rad)
    voy = v0 * np.sin(theta_rad)
    t_flight = (voy + np.sqrt(voy**2 + 2*g*y0)) / g
    if t_flight < 0:
        t_flight = (voy - np.sqrt(voy**2 + 2*g*y0)) / g


    t = np.linspace(0, t_flight, 100)


    x = x0 + vox * t
    y = y0 + voy * t - 0.5 * g * t**2


    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=f'Projectile Trajectory (v0={v0} m/s, $\theta$={theta_deg})')
    plt.xlabel('Horizontal Distance (m)')
    plt.ylabel('Vertical Distance (m)')
    plt.title('Projectile Motion Simulation')
    plt.grid(True)
    plt.legend()
    plt.axhline(0, color='black', linestyle='--', linewidth=0.8) # Ground line
    plt.ylim(bottom=0)
    plt.show()


projectile_motion_simulation()
```

```python
import numpy as np
def mat_multi_numpy(Alist,Blist):
   # Args:
   #    Alist(list[[int/float]]):the first matrix
   #    Blist(list[[int/float]]):the second matrix
   # Returns:
   #    numpy.ndarray: The resulting matrix
   A=np.array(Alist)
   B=np.array(Blist)
   C=A@B
   return C
```

```python
import sympy

# Define symbolic variables
x, y, z = sympy.symbols('x y z')

# Define the vector field G
G = sympy.Matrix([y**2 + z, -x + z**2, x**2 + y])

# Calculate the curl
curl_G = sympy.Matrix([
    sympy.diff(G[2], y) - sympy.diff(G[1], z),  # i component
    sympy.diff(G[0], z) - sympy.diff(G[2], x),  # j component
    sympy.diff(G[1], x) - sympy.diff(G[0], y)   # k component
])

# Simplify the curl
curl_G = sympy.simplify(curl_G)

# Print the curl
print("Curl of G:")
sympy.pprint(curl_G)

# You can also get individual components:
curl_i = curl_G[0]
curl_j = curl_G[1]
curl_k = curl_G[2]
```

```python
import sympy

# Define symbolic variables
x, y, z = sympy.symbols('x y z')

# Define the scalar function phi
phi = x**2 + y + y * z**2

# Calculate the gradient
gradient_phi = sympy.Matrix([
    sympy.diff(phi, x),  # Partial derivative with respect to x
    sympy.diff(phi, y),  # Partial derivative with respect to y
    sympy.diff(phi, z)   # Partial derivative with respect to z
])

# Simplify the gradient
gradient_phi = sympy.simplify(gradient_phi)

# Print the gradient
print("Gradient of phi:")
sympy.pprint(gradient_phi)

# Evaluate the gradient at the point (1, 2, -1)
point = {x: 1, y: 2, z: -1}
gradient_at_point = gradient_phi.subs(point)
print("\nGradient at point (1, 2, -1):")
sympy.pprint(gradient_at_point)
```

```python
import sympy

# Define symbolic variables
x, y, z = sympy.symbols('x y z')

# Define the vector field G (example)
G = sympy.Matrix([y**2 + z, -x + z**2, x**2 + y])

# Calculate the divergence
divergence_G = sympy.diff(G[0], x) + sympy.diff(G[1], y) + sympy.diff(G[2], z)
divergence_G = sympy.simplify(divergence_G)

# Print the divergence
print("Divergence of G:")
sympy.pprint(divergence_G)

# Evaluate the divergence at the point (1, 2, -1)
point = {x: 1, y: 2, z: -1}
divergence_at_point = divergence_G.subs(point)
print("\nDivergence at point (1, 2, -1):")
sympy.pprint(divergence_at_point)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def projectile_motion_simulation():
    # Simulates projectile motion and plots multiple trajectories

    # Get user inputs for common parameters
    v0 = float(input("Enter initial velocity (v0) in m/s: "))
    y0 = float(input("Enter initial vertical position (y0) in meters: "))
    x0 = float(input("Enter initial horizontal position (x0) in meters: "))

    # Get user inputs for angle range and step
    start_angle_deg = float(input("Enter starting launch angle in degrees: "))
    end_angle_deg = float(input("Enter ending launch angle in degrees: "))
    angle_step_deg = float(input("Enter angle step in degrees: "))

    # Constants
    g = 9.81 # Acceleration due to gravity (m/s^2)

    plt.figure(figsize=(10, 6))

    # Iterate through the angles
    for theta_deg in np.arange(start_angle_deg, end_angle_deg + angle_step_deg, angle_step_deg):
        if theta_deg > 90 or theta_deg < -90: # Simple validation for reasonable angles
            print(f"Skipping angle {theta_deg}: Angle out of practical range.")
            continue

        theta_rad = np.radians(theta_deg)

        # Calculate initial velocity components
        vox = v0 * np.cos(theta_rad)
        voy = v0 * np.sin(theta_rad)

        # Calculate time of flight (assuming landing at y=0, solving quadratic equation)
        # 0 = y0 + voy*t - 0.5*g*t^2
        # 0.5*g*t^2 - voy*t - y0 = 0
        # Using quadratic formula: t = (-b +/- sqrt(b^2 - 4ac)) / 2a
        # a = 0.5*g, b = -voy, c = -y0
        discriminant = voy**2 + 2 * g * y0
        if discriminant < 0:
            print(f"Error for angle {theta_deg}: Projectile does not hit the ground.")
            continue
        t_flight = (voy + np.sqrt(discriminant)) / g
        # Ensure t_flight is positive
        if t_flight < 0:
            t_flight = (voy - np.sqrt(discriminant)) / g

        # Time array for plotting
        t = np.linspace(0, t_flight, 100)

        # Calculate x and y positions
        x = x0 + vox * t
        y = y0 + voy * t - 0.5 * g * t**2

        # Calculate horizontal range (R)
        R = x0 + vox * t_flight # x position at time of flight

        # Plotting for current angle
        plt.plot(x, y, label=f'$\theta$={theta_deg} (R={R:.2f}m)')

        # Print calculations for current angle
        print(f"\nCalculations for angle {theta_deg}:")
        print(f"Horizontal Range (R): {R:.2f} meters")
        print(f"Time of Flight: {t_flight:.2f} seconds")

    # Final plot settings
    plt.xlabel('Horizontal Distance (x) in meters')
    plt.ylabel('Vertical Distance (y) in meters')
    plt.title('Projectile Motion Trajectory') # Updated title as per user request
    plt.grid(True)
    plt.axhline(0, color='black', linestyle='--', linewidth=0.8) # Ground line
    plt.ylim(bottom=0) # Ensure y-axis starts from 0 or above
    plt.legend()
    plt.show()

# Run the simulation
projectile_motion_simulation()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def calculate_accelerations(angle_deg, g=9.81):
    angle_rad = np.radians(angle_deg)
    accelerations = {
        "spherical shell": (3/5) * g * np.sin(angle_rad),
        "solid sphere": (5/7) * g * np.sin(angle_rad),
        "cylindrical shell": (1/2) * g * np.sin(angle_rad),
        "solid cylinder": (2/3) * g * np.sin(angle_rad),
        "sliding cube": g * np.sin(angle_rad),
    }
    return accelerations

def plot_accelerations_bar(angle_deg):
    accelerations = calculate_accelerations(angle_deg)
    objects = list(accelerations.keys())
    acc_values = list(accelerations.values())

    plt.figure(figsize=(10, 6))
    plt.bar(objects, acc_values)
    plt.xlabel("Object Type")
    plt.ylabel("Acceleration (m/s^2)")
    plt.title(f"Acceleration of Objects at {angle_deg} Incline")
    plt.grid(True) # This is the corrected line.
    plt.show()

# Get user input for angle
angle_deg = float(input("Enter incline angle (degrees): "))
plot_accelerations_bar(angle_deg)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def simulate_collision(m1, v1i, m2, v2i, e):
    # Using the restitution formula and conservation of momentum
    v1f = (v1i * (m1 - e * m2) + v2i * m2 * (1 + e)) / (m1 + m2)
    v2f = (v1i * m1 * (1 + e) + v2i * (m2 - e * m1)) / (m1 + m2)
    return v1f, v2f

def plot_collision(m1, v1i, m2, v2i, e):

    v1f, v2f = simulate_collision(m1, v1i, m2, v2i, e)

    time = np.linspace(0, 10, 100) # Time range
    x1_initial = 5  # Initial position of mass 1
    x2_initial = 15 # Initial position of mass 2

    x1_before = x1_initial + v1i * time
    x2_before = x2_initial + v2i * time

    x1_after = x1_initial + v1f * time
    x2_after = x2_initial + v2f * time

    plt.figure(figsize=(12, 6))

    # Position Plot
    plt.subplot(1, 2, 1)
    plt.plot(time, x1_before, 'r--', label=f'Mass 1 (Before Collision, v={v1i}m/s)')
    plt.plot(time, x2_before, 'b--', label=f'Mass 2 (Before Collision, v={v2i}m/s)')
    plt.plot(time, x1_after, 'r-', label=f'Mass 1 (After Collision, v={v1f:.2f}m/s)')
    plt.plot(time, x2_after, 'b-', label=f'Mass 2 (After Collision, v={v2f:.2f}m/s)')

    plt.xlabel("Time (s)")
    plt.ylabel("Position (m)")
    plt.title(f"Position vs Time (e={e})")
    plt.legend()
    plt.grid(True)

    # Velocity Plot
    plt.subplot(1, 2, 2)
    plt.axhline(v1i, color='r', linestyle='--', label=f'Mass 1 (Before Collision, v={v1i}m/s)')
    plt.axhline(v2i, color='b', linestyle='--', label=f'Mass 2 (Before Collision, v={v2i}m/s)')
    plt.axhline(v1f, color='r', label=f'Mass 1 (After Collision, v={v1f:.2f}m/s)')
    plt.axhline(v2f, color='b', label=f'Mass 2 (After Collision, v={v2f:.2f}m/s)')

    plt.xlabel('Time (s)')
    plt.ylabel('Velocity (m/s)')
    plt.title('Velocity Before and After Collision')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    print(f"Mass 1 initial velocity: {v1i:.2f} m/s, final velocity: {v1f:.2f} m/s")
    print(f"Mass 2 initial velocity: {v2i:.2f} m/s, final velocity: {v2f:.2f} m/s")

# Example usage:
# m1: mass of object 1, m2: mass of object 2 (kg)
# v1i: initial velocity of object 1, v2i: initial velocity of object 2 (m/s)
# e: coefficient of restitution (0 for perfectly inelastic, 1 for perfectly elastic)
print("\n--- Elastic collision (m1=2, v1i=5, m2=1, v2i=0, e=1) ---")
plot_collision(m1=2, v1i=5, m2=1, v2i=0, e=1) # Elastic collision

print("\n--- Perfectly inelastic collision (m1=2, v1i=5, m2=1, v2i=0, e=0) ---")
plot_collision(m1=2, v1i=5, m2=1, v2i=0, e=0) # Perfectly inelastic collision

print("\n--- General case (m1=2, v1i=5, m2=1, v2i=-3, e=0.5) ---")
plot_collision(m1=2, v1i=5, m2=1, v2i=-3, e=0.5) # General case
```
